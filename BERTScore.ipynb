```python
import transformers
import pandas as pd
from time import perf_counter
from bert_score import score
```

```python
# get candidates and references (from https://github.com/Tiiiger/bert_score/tree/master/example)
with open("hyps_long.txt") as f:
    cands = [line.strip() for line in f]

with open("refs_long.txt") as f:
    refs = [line.strip() for line in f]
```

Calculating scores

```python
VARIANT = 'Variant'
F1SCORE = 'F1 Score'
TIME = 'Time'
BERTTIME = '% Time BERT'
```

```python
# build dictionary and set alternatives to be tested
variants = ["bert-base-uncased", "roberta-base", "albert-base-v2", "distilroberta-base", "microsoft/deberta-base"]
results = {
    VARIANT: [],
    F1SCORE: [],
    TIME: [],
    BERTTIME: []*5
}
```

```python
# TODO run multiple times to get more robust averages 
# calculate f1 scores + durations for all variants
F1_scores = []
t = []
for v in variants:
  results[VARIANT].append(v)
  s = perf_counter()
  Precision, Recall, F1 = score(cands, refs, lang='en', model_type=v)
  results[TIME].append(perf_counter() - s)
  results[F1SCORE].append(f'{F1.mean():.3f}')
```

```python
# compare time taken to base bert time
results[BERTTIME] = []
for i in range(len(variants)):
  results[BERTTIME].append(f'{results[TIME][i] / results[TIME][0] * 100:.1f}')
```

```python
# build and save dataframe
df = pd.DataFrame(results, columns=[VARIANT, F1SCORE, TIME, BERTTIME]).set_index(VARIANT)
df.to_csv("data.csv", index=False)
```
